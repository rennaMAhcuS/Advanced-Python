### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plots

### Scatter Plot

```python
X_data_scatter_1 = np.random.random(50) * 100
Y_data_scatter_1 = np.random.random(50) * 100
plt.scatter(X_data_scatter_1, Y_data_scatter_1)
```

```python
X_data_scatter_2 = np.random.random(500) * 100
Y_data_scatter_2 = np.random.random(500) * 100
plt.scatter(X_data_scatter_2, Y_data_scatter_2,
            color='#007BFF',
            s=80,  # Size
            marker='*',  # The datapoint marker
            alpha=0.2  # Transparency
            )
```

### Line Graph

```python
years_line = np.arange(2006, 2022)
weights_line = np.random.randint(60, 80, size=16)
plt.plot(years_line, weights_line,
         color='red',  # Line colour
         linestyle='--',
         linewidth=3
         )
```

### Bar Chart

```python
x_bar = np.array(["C++", "C#", "Python", "Java", "Go"])
y_bar = np.array([20, 50, 140, 5, 45])
plt.bar(x_bar, y_bar,
        color='red',
        align='center',  # The bar w.r.t the X - marker,
        width=0.5,
        edgecolor='lightblue', # Border
        linewidth=3  # Border
        )
```

### Histogram

```python
ages_hist = np.random.normal(25, 2.5, 1000)
plt.hist(ages_hist,
         bins=40,  # Also supports an array
         color='red'
         )
```

```python
# Cumulative Graph
plt.hist(ages_hist,
         bins=40,  # Also supports an array
         cumulative=True
         )
```

### Pie Chart

```python
langs_pie = ['Python', 'C++', 'Java', 'C#', 'Go']
votes_pie = [50, 24, 14, 6, 17]
explodes_pie = [0.1, 0, 0, 0.2, 0]
plt.pie(votes_pie,
        labels=langs_pie,
        explode=explodes_pie,  # The distance coming out
        autopct="%.1f%%",  # The format of the displayed %
        pctdistance=0.8,  # Distance of the displayed %
        startangle=90  # Starting angle position (Similar to the Cartesian plane)
        )
```

## Customizations

```python
# Labeling the axes

years = np.arange(2014, 2022)
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = np.arange(50, 81, 2)

plt.plot(years, income)
plt.title("Income of John (in USD)",
          fontsize=20,  # Font Size
          fontname='Comic Sans MS'  # Font Name (Family)
          )
plt.xlabel ("Year")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f'${x}K' for x in income_ticks])  # The markers on the Y-axis

# Note: This customization can also be applied for the others (Bar, Pie, Scatter, etc)
```

```python
# Introducing a legend

stock_a = [100, 102, 99, 101, 101, 100, 102]
stock_b = [90, 95, 102, 104, 105, 103, 109]
stock_c = [110, 115, 100, 105, 100, 98, 95]

plt.plot(stock_a, label="Company 1")
plt.plot(stock_b, label="Company 2" )
plt.plot(stock_c, label="Company 3")

plt.legend(loc="lower center")
```

```python
votes = [10, 2, 5, 16, 22]
people = ["A", "B", "C", "D", "E"]

plt.pie(votes, labels=None)
plt.legend(labels=people, loc="lower right")
```

```python
from matplotlib import style
```

```python
# Style (themes)

style.use("dark_background")

plt.pie(votes, labels=None)
plt.legend(labels=people, loc="lower right")
```

## Subplots and Saving the Plot

```python
x = np.arange(1, 360, 3) * np.pi / 180

fig, axs = plt.subplots(2, 2)
fig.suptitle("Many Plots!")

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(len(x)))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log Function")

plt.tight_layout()

plt.savefig('FourPlots.png', dpi=360, transparent=False)
```
