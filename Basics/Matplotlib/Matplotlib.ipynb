# Matplotlib (Source: NeuralNine)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data_scatter_1 = np.random.random(50) * 100
Y_data_scatter_1 = np.random.random(50) * 100
plt.scatter(X_data_scatter_1, Y_data_scatter_1)
```

```python
X_data_scatter_2 = np.random.random(500) * 100
Y_data_scatter_2 = np.random.random(500) * 100
plt.scatter(X_data_scatter_2, Y_data_scatter_2,
            c='#007BFF',  # Colour
            s=80,  # Size
            marker='*',  # The datapoint marker
            alpha=0.2  # Transparency
            )
```

```python
years_line = np.arange(2006, 2022)
weights_line = np.random.randint(60, 80, size=16)
plt.plot(years_line, weights_line,
         c='red',  # Line colour
         ls='--',  # Line style
         lw=3,  # Line width
         )
```

```python
x_bar = np.array(["C++", "C#", "Python", "Java", "Go"])
y_bar = np.array([20, 50, 140, 5, 45])

plt.bar(x_bar, y_bar,
        color='r',
        align='center',  # The bar w.r.t the X - marker,
        width=0.5
        )
```
